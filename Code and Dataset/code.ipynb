```python
#mapping categorical variables to numericalimport pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv('personality_synthetic_dataset.csv')
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].astype('category').cat.codes
print("\nData types after encoding categorical variables:")
print(df.dtypes)
```

```python
#from here data analysis starts
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split

#separate features and target variable
x = df.drop('personality_type', axis=1)
y = df['personality_type']
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)

#initialize PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
pca_df = pd.DataFrame(data=x_pca, columns=['PC1', 'PC2'])
pca_df['personality_type'] = y.values

# Plotting the PCA results
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='personality_type', palette='viridis', alpha=0.7)
plt.title('PCA of Personality Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Personality Type')
plt.grid()
plt.show()
print("\nPCA Results:")
print(pca_df.head().round(3))
```

```python
#random forest classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score

# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(x_train, y_train)

y_pred = rf_classifier.predict(x_test)
print("Random Forest Classifier Results:")
print("="* 50)

print("\nclassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print(f"\nAccuracy: {accuracy_score(y_test, y_pred):.4f}")

# Feature Importance
feature_importances = rf_classifier.feature_importances_
importance_df = pd.DataFrame({'Feature': x.columns, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
print("\nFeature Importance (Random Forest):")
print(importance_df)
```

```python
#Putting threshold of >=0.25 for classification(doubt)

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Encode target labels
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)

# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(x_train, y_train_encoded)
y_pred_proba = rf_classifier.predict_proba(x_test)

# Apply threshold of 0.25 for each class
threshold = 0.25
y_pred = []
for probs in y_pred_proba:
	above_thresh = np.where(probs >= threshold)[0]
	if len(above_thresh) == 1:
		y_pred.append(above_thresh[0])
	elif len(above_thresh) > 1:
		y_pred.append(above_thresh[np.argmax(probs[above_thresh])])
	else:
		y_pred.append(np.argmax(probs))
y_pred = np.array(y_pred)

y_pred_labels = le.inverse_transform(y_pred)

print("Random Forest Classifier Results (with threshold >= 0.25):")
print("=" * 50)

print("\nClassification Report:")
print(classification_report(y_test, y_pred_labels))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_labels))
print(f"\nAccuracy: {accuracy_score(y_test, y_pred_labels):.4f}")

feature_importances = rf_classifier.feature_importances_
importance_df = pd.DataFrame({'Feature': x.columns, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Plotting feature importances
plt.figure(figsize=(12, 6))
sns.barplot(data=importance_df, x='Importance', y='Feature', palette='viridis')
plt.title('Feature Importances from Random Forest Classifier')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.grid()
plt.show()
```

```python
#k-NN with threshold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train, y_train)
y_probs = knn_model.predict_proba(x_test)
threshold = 0.5 

# Apply threshold
y_pred_knn = []
for prob in y_probs:
    if max(prob) > threshold:
        y_pred_knn.append(knn_model.classes_[np.argmax(prob)])
    else:
        y_pred_knn.append("Undetermined") 

print("k-Nearest Neighbors with Threshold Results:")
print("=" * 50)
print("\nClassification Report:")
print(classification_report(y_test, y_pred_knn))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_knn))

# Convert y_pred_knn to numpy array for boolean indexing(understanding uncertainty)
y_pred_knn_arr = np.array(y_pred_knn)
mask = y_pred_knn_arr != 'Undetermined'
print(f"\nAccuracy (excluding uncertain): {accuracy_score(y_test[mask], y_pred_knn_arr[mask]):.4f}")
print(f"Uncertainty Rate: {(y_pred_knn_arr == 'Undetermined').sum()/len(y_pred_knn_arr):.4f}")

#plotting in horizontal barplot
plt.figure(figsize=(10, 6))
sns.countplot(y=y_pred_knn_arr, order=pd.Series(y_pred_knn_arr).value_counts().index, palette='viridis')
plt.title('k-NN Predictions with Threshold')
plt.xlabel('Count')
plt.ylabel('Predicted Personality Type')
plt.grid()
plt.show()
```

```python
# knn with manhattan distance (ambivert classification)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
le_knn = LabelEncoder()
y_train_knn = le_knn.fit_transform(y_train)
y_test_knn = le_knn.transform(y_test)

knn_manhattan = KNeighborsClassifier(n_neighbors=5, metric='manhattan')
knn_manhattan.fit(x_train, y_train_knn)
y_probs_knn_manhattan = knn_manhattan.predict_proba(x_test)

threshold = 0.5
y_pred_knn_manhattan = []
for prob in y_probs_knn_manhattan:
	if max(prob) > threshold:
		y_pred_knn_manhattan.append(np.argmax(prob))
	else:
		y_pred_knn_manhattan.append(-1) 
y_pred_knn_manhattan = np.array(y_pred_knn_manhattan)

y_pred_knn_manhattan_labels = []
for pred in y_pred_knn_manhattan:
	if pred == -1:
		y_pred_knn_manhattan_labels.append('Undetermined')
	else:
		y_pred_knn_manhattan_labels.append(le_knn.inverse_transform([pred])[0])
y_pred_knn_manhattan_labels = np.array(y_pred_knn_manhattan_labels)

print("\nClassification Report:")
print(classification_report(y_test, y_pred_knn_manhattan_labels))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_knn_manhattan_labels))
print(f"\nAccuracy: {accuracy_score(y_test, y_pred_knn_manhattan_labels):.4f}")

# Convert y_pred_knn_manhattan_labels to numpy array for boolean indexing
y_pred_knn_manhattan_arr = np.array(y_pred_knn_manhattan_labels)
mask_manhattan = y_pred_knn_manhattan_arr != 'Undetermined'
print(f"\nAccuracy (excluding uncertain): {accuracy_score(y_test[mask_manhattan], y_pred_knn_manhattan_arr[mask_manhattan]):.4f}")
print(f"Uncertainty Rate: {(y_pred_knn_manhattan_arr == 'Undetermined').sum()/len(y_pred_knn_manhattan_arr):.4f}")
```
